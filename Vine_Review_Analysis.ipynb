```python
import pandas as pd
import os
```

```python
vine_data = os.path.join('vine_table.csv')
vine_df = pd.read_csv(vine_data)
```

```python
vine_df.head(10)
```

```python
# Step 1
new_df=vine_df[vine_df['total_votes'] > 19]
new_df.head(10)
```

```python
#Step 2.a
new_df['results'] = new_df['helpful_votes'] / new_df['total_votes']
new_df.head(10)
```

```python
#Step 2.b
new_df2 = new_df[new_df['results'] > 0.5]
new_df2.head(10)
```

```python
# Step 3
paid_df = new_df2[new_df2['vine'] == 'Y']
paid_df.head(10)
```

```python
# Step 4
unpaid_df = new_df2[new_df2['vine'] == 'N']
unpaid_df.head(10)
```

```python
# count of paid Vine votes
paid_df['total_votes'].count()
```

```python
# count of unpaid Vine votes
unpaid_df['total_votes'].count()
```

```python
#number of paid Vine 5-star reviews
len(paid_df[paid_df['star_rating']==5])
```

```python
#number of unpaid Vine 5-star reviews
len(unpaid_df[unpaid_df['star_rating']==5])
```

```python
#Percent 5 star review paid Vine
(len(paid_df[paid_df['star_rating']==5]))/(paid_df['total_votes'].count())
```

```python
#Percent 5 star review unpaid Vine
(len(unpaid_df[unpaid_df['star_rating']==5]))/(unpaid_df['total_votes'].count())
```
